```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how = "outer", on="Mouse ID")
# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
mice_count = len(combined_df["Mouse ID"].unique())
print(f'There are {mice_count} mice in the animal study')
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_df = pd.DataFrame(combined_df.drop_duplicates(subset=["Mouse ID", "Timepoint"]))
clean_combined_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count1 = len(combined_df["Mouse ID"].unique())
print(f'Even after dropping some data, there are still {mice_count1} mice in the animal study')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
stats = clean_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(['mean','median', 'var', 'std', 'sem'])
stats
# This method produces everything in a single groupby function
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
mouse_count = clean_combined_df.groupby(["Drug Regimen"])["Mouse ID"].count()

mouse_count.plot(kind="bar",color = "red", figsize=(6,4))
plt.title("Number of Mice per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice")
plt.xlim(-1,10)
plt.ylim(0,250)
```

## Quartiles, Outliers and Boxplots

```python
filtered_df = clean_combined_df.loc[(clean_combined_df["Drug Regimen"] == "Capomulin") | (clean_combined_df["Drug Regimen"] == "Ramicane") | (clean_combined_df["Drug Regimen"] == "Infubinol") | (clean_combined_df["Drug Regimen"] == "Ceftamin"), :]

#Sort by Timepoints and then keeping the first value, which will give us our last turmor volume
filtered_df = filtered_df.sort_values("Timepoint", ascending = False)

# Dropping duplicates, keeping first value, should be the latest timepoint per mouse
filtered_df = filtered_df.drop_duplicates(subset="Mouse ID", keep='first')
filtered_df
```

```python
capomulin = filtered_df.loc[(clean_combined_df["Drug Regimen"]=="Capomulin"),"Tumor Volume (mm3)"]
ramicane = filtered_df.loc[(clean_combined_df["Drug Regimen"]=="Ramicane"),"Tumor Volume (mm3)"]
infubinol = filtered_df.loc[(clean_combined_df["Drug Regimen"]=="Infubinol"),"Tumor Volume (mm3)"]
ceftamin = filtered_df.loc[(clean_combined_df["Drug Regimen"]=="Ceftamin"),"Tumor Volume (mm3)"]
```

```python
#Setting the values for the loop
drug_name = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
drug_l = [capomulin,ramicane,infubinol,ceftamin]
x = 0

for drug in drug_l:
    quartiles = drug.quantile([.25,.5,.75])
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    print(f'---------------------------------')
    print(f"The lower quartile of {drug_name[x]} is: {lowerq}")
    print(f"The upper quartile of {drug_name[x]} is: {upperq}")
    print(f"The interquartile range of {drug_name[x]} is: {iqr}")
    print(f"The median of {drug_name[x]} is: {quartiles[0.5]} ")
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    print(f'--------------------------------------')
    x=x+1
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
drug_l = [capomulin,ramicane,infubinol,ceftamin]
drug_name = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
red_diamond = dict(markerfacecolor='r', marker='D')
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume of Each Mouse')
ax1.set_ylabel('Tumor Vol (mm3)')
ax1.set_xlabel('Drug Regimen')
ax1.boxplot(drug_l, flierprops=red_diamond)
plt.xticks([1,2,3,4],drug_name)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

#Finding the information for Mouse s185
mouse1 = clean_combined_df.loc[(clean_combined_df["Drug Regimen"]=="Capomulin"),:]
mouse1 = mouse1.loc[(mouse1["Mouse ID"]=="s185"),:]
mouse1
```

```python
#Generating the line plot of time point versus tumor volume for mouse s185
s185_handle, = plt.plot(mouse1["Timepoint"], mouse1["Tumor Volume (mm3)"], marker="o", color="blue")
plt.title("Time Point vs. Tumor Volume (mm3)")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
mice1 = clean_combined_df.loc[(clean_combined_df["Drug Regimen"]=="Capomulin"),:]

weight = mice1.groupby(mice1["Mouse ID"])["Weight (g)"].mean()
volume = mice1.groupby(mice1["Mouse ID"])["Tumor Volume (mm3)"].mean()

# Create Scatter Plot with values calculated above
plt.scatter(weight,volume)
plt.title("Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
plt.xticks(weight)
plt.show()
```

## Correlation and Regression


```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

weight = mice1.groupby(mice1["Mouse ID"])["Weight (g)"].mean()
volume = mice1.groupby(mice1["Mouse ID"])["Tumor Volume (mm3)"].mean()

# Perform a linear regression on tumor volume versus weight
slope, int, r, p, std_err = st.linregress(weight,volume)
      
# Create equation of line to calculate predicted weight
fit = slope * weight + int

# Plot the linear model on top of scatter plot 
plt.scatter(weight,volume)
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
plt.plot(weight,fit,"--")
plt.xticks(weight)
plt.show()

# Caculate correlation coefficient
corr = round(st.pearsonr(weight,volume)[0],2)
print(f'The correlation between weight and tumor value is {corr}')
```
